```python
%load_ext autoreload
%autoreload 2

from cs336_data.parsing import extract_text_from_html_bytes
import gzip

DATA_FILE = "../data/CC-MAIN-20180420081400-20180420101400-00118.warc.gz"
WET_FILE = "../data/CC-MAIN-20180420081400-20180420101400-00118.warc.wet.gz"
with gzip.open(DATA_FILE) as file:
    unparsed = file.read(50000)

parsed = extract_text_from_html_bytes(unparsed)
print(parsed)
```

### 2.3 Language Identification

```python
from warcio.archiveiterator import ArchiveIterator
from cs336_data.language_id import identify_language

num_iters = 20

with open(WET_FILE, "rb") as stream:
    for idx, record in enumerate(ArchiveIterator(stream)):
        if idx == 0:
            continue
        content = record.content_stream().read()
        content = content.decode()
        print(f"Entry no: {idx}")
        print(content[0:200])
        # print(content)
        result = identify_language(content)
        print(result)
        if idx == num_iters:
            break
        print("\n\n\n")
```

### 2.4 Personal Identifiable Information

```python
import itertools
from warcio.archiveiterator import ArchiveIterator
from cs336_data.privacy import mask_emails, mask_ips, mask_emails

num_iters = 20
removed_count = 0

masks = [mask_emails, mask_ips, mask_emails]

with open(WET_FILE, "rb") as stream:
    iterator = itertools.islice(ArchiveIterator(stream), 0, None)
    for idx, record in enumerate(iterator):
        if idx == 0:
            continue
        content = record.content_stream().read()
        text = extract_text_from_html_bytes(content)
        total_removed = 0
        for mask in masks:
            text, removed = mask(text)
            total_removed += removed
        if total_removed > 0:
            removed_count += 1
            print(f"Entry no: {idx}")   
            print(text)
            print(f"Removed: {total_removed}")
            print("NEW ENTRY\n\n")

        if removed_count == num_iters:
            break
```

### 2.5 Harmful Content

```python
import itertools
from warcio.archiveiterator import ArchiveIterator
from cs336_data.harmful import classify_nsfw, classify_toxic_speech

num_iters = 20
count = 0
start_idx = 1020

with open(WET_FILE, "rb") as stream:
    iterator = itertools.islice(ArchiveIterator(stream), start_idx, None)
    for idx, record in enumerate(iterator, start_idx):
        if idx == 0:
            continue
        content = record.content_stream().read()
        text = extract_text_from_html_bytes(content)
        toxic_label = classify_toxic_speech(text)
        nsfw_label = classify_nsfw(text)

        print(f"Entry No. {idx}")
        print(f"Toxic: {toxic_label}")
        print(f"NSFW: {nsfw_label}")
        print(text)
        print("\n\n\n")
        count += 1

        if count == num_iters:
            break
```

### 2.6 Quality Rules

```python
import itertools
from warcio.archiveiterator import ArchiveIterator
from cs336_data.quality import gopher_quality_filter

num_iters = 20
count = 0
start_idx = 2010

with open(WET_FILE, "rb") as stream:
    iterator = itertools.islice(ArchiveIterator(stream), start_idx, None)
    for idx, record in enumerate(iterator, start_idx):
        if idx == 0:
            continue
        content = record.content_stream().read()
        text = extract_text_from_html_bytes(content)
        quality_label = gopher_quality_filter(text)

        print(f"Entry No. {idx}")
        print(f"Quality: {quality_label}")
        print(text)
        print("\n\n\n")
        count += 1

        if count == num_iters:
            break
```
